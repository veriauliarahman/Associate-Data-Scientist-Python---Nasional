```python
# 1. KOMENTAR: Ini adalah bagian untuk mencatat data transaksi pelanggan
nama_pelanggan = "Rian"      # Tipe data: String (str) untuk teks
harga_kopi = 25000.0         # Tipe data: Float (float) untuk angka pecahan/desimal
jumlah_gelas = 3             # Tipe data: Integer (int) untuk angka bulat
is_member = True             # Tipe data: Boolean (bool) untuk Benar/Salah
```

```python
# 2. OPERATOR ARITMATIKA: Menghitung total harga kopi yang dibeli
total_belanja = harga_kopi * jumlah_gelas

print("Nama Pelanggan:", nama_pelanggan)
print("Total yang harus dibayar awal: Rp", total_belanja)
```

```python
# 3. OPERATOR PERBANDINGAN & LOGIKA
# Syarat 1: Apakah dia membeli lebih dari 2 gelas?
beli_banyak = jumlah_gelas > 2  # Menghasilkan True karena 3 > 2

# Syarat 2: Apakah syarat 1 DAN status member-nya dua-duanya terpenuhi?
dapat_diskon = beli_banyak and is_member

print("Apakah", nama_pelanggan, "berhak mendapat diskon?", dapat_diskon)
```

```python
# 4. IF-ELIF-ELSE: Menentukan persentase diskon
# Skenario: Member beli banyak dikasih 15%, member biasa 5%, selain itu 0%

if is_member and jumlah_gelas > 2:
    persen_diskon = 0.15  # Diskon 15%
elif is_member:
    persen_diskon = 0.05  # Diskon 5%
else:
    persen_diskon = 0.0   # Tidak ada diskon

# Menghitung potongan dan total akhir
potongan_harga = total_belanja * persen_diskon
total_akhir = total_belanja - potongan_harga

print(f"Diskon yang didapat: {persen_diskon * 100}%")
print("Total yang harus dibayar Rian: Rp", total_akhir)
```

```python
# 5. FOR LOOP: Mengabsen daftar pesanan di dalam List
menu_dipesan = ["Espresso", "Caffè Latte", "Cappuccino"]

print("Detail Kopi yang Dipesan:")
# Variabel 'kopi' akan bergantian mengambil isi dari 'menu_dipesan'
for kopi in menu_dipesan:
    print("- " + kopi)
```

```python
# 6. FUNGSI: Membuat cetakan kasir otomatis yang fleksibel
def hitung_kasir_digital(nama, harga, jumlah, status_member):
    # Proses Perhitungan (Logika yang kita pelajari tadi)
    total = harga * jumlah
    
    if status_member and jumlah > 2:
        diskon = 0.15
    elif status_member:
        diskon = 0.05
    else:
        diskon = 0.0
        
    total_akhir = total - (total * diskon)
    
    # Return: Mengembalikan hasil akhir cetakan ke luar fungsi
    return f"Pelanggan: {nama} | Total Bayar: Rp {total_akhir}"

# Sekarang mari kita panggil fungsi ini untuk pelanggan yang berbeda-beda!
Struk_1 = hitung_kasir_digital("Alice", 30000.0, 4, True)
Struk_2 = hitung_kasir_digital("Budi", 20000.0, 1, False)

print(Struk_1)
print(Struk_2)
```

```python
import json

# --- 1. STRUKTUR DATA (Dictionary & List) ---
# Kita simpan stok gudang kopi menggunakan Dictionary (Key-Value)
stok_gudang = {
    "Espresso": 50,
    "Cappuccino": 30,
    "Latte": 25
}

# --- 2. ERROR HANDLING (Jaring Pengaman) ---
print("--- PROSES PENGURANGAN STOK ---")
menu_diambil = "Espresso"
jumlah_ambil = "10" # Ops! sengaja kita buat salah (harus angka, tapi diisi teks)

try:
    # Komputer akan mencoba mengubah teks menjadi angka bulat (int)
    angka_ambil = int(jumlah_ambil) 
    stok_gudang[menu_diambil] -= angka_ambil
    print(f"Stok {menu_diambil} berhasil dikurangi. Sisa stok: {stok_gudang[menu_diambil]}")
    
except ValueError:
    # Jaring pengaman bekerja jika terjadi salah input tipe data
    print("❌ ERROR: Kamu salah memasukkan jumlah! Harus berupa angka (misal: 10), bukan teks.")
    print("Gagal memproses... Tapi program kita AMAN dan tidak crash!\n")

# --- 3. FILE HANDLING (Menulis data ke laptop) ---
print("--- MENYIMPAN LAPORAN GUDANG ---")
nama_file = "laporan_gudang.json"

try:
    # Kita buka file dalam mode "w" (Write) untuk menyimpan stok terbaru
    with open(nama_file, "w") as file:
        # Mengubah dictionary menjadi format teks JSON agar rapi di file
        json.dump(stok_gudang, file, indent=4) 
    print(f"✅ Hore! Laporan stok berhasil disimpan secara fisik di laptopmu dengan nama: '{nama_file}'")
    
except Exception as e:
    print(f"❌ Gagal menyimpan file karena: {e}")
```
