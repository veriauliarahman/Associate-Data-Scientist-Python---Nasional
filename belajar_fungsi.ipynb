```python
# 1. MENGGUNAKAN DOCSTRING & ARGUMEN DEFAULT
def sapa_siswa(nama="Siswa Misterius"):
    """
    Fungsi ini digunakan untuk menyapa siswa.
    Jika nama tidak diisi, otomatis akan memanggil 'Siswa Misterius'.
    """
    return f"Halo, {nama}! Mari kita analisis nilaimu."

# 2. MENGGUNAKAN *ARGS & FUNGSI BAWAAN
def hitung_rata_rata(*nilai):
    """Menghitung rata-rata dari berapapun jumlah nilai yang dimasukkan."""
    total = sum(nilai)            # sum() adalah fungsi bawaan Python
    jumlah_data = len(nilai)      # len() juga fungsi bawaan Python
    return total / jumlah_data

# 3. MENGGUNAKAN LAMBDA (FUNGSI SATU BARIS)
# Kriteria lulus: Rata-rata harus 70 atau lebih
cek_lulus = lambda rata_rata: "Lulus 🎉" if rata_rata >= 70 else "Remedial 📚"

# ==========================================
# --- MARI KITA JALANKAN ---
# ==========================================

# Kita panggil tanpa mengisi nama (Menguji Argumen Default)
print(sapa_siswa())

# Kita panggil dengan nama asli
print(sapa_siswa("Budi"))

# Menguji *args (Budi memasukkan 4 nilai ujian sekaligus)
rata_budi = hitung_rata_rata(80, 75, 90, 65)
print(f"Rata-rata Budi: {rata_budi}")
print(f"Status Budi: {cek_lulus(rata_budi)}")

print("\n" + "="*30)
print("MEMANGGIL BANTUAN (HELP):")
# Kita minta Python menjelaskan cara kerja fungsi yang kita buat!
help(hitung_rata_rata)
```
