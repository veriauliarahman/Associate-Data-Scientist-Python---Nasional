```python
import random

# --- 1. CLASS (Cetakan Dasar) ---
# Kita buat cetakan makhluk hidup. Semua makhluk pasti punya nama, nyawa (HP), dan kekuatan serang.
class Karakter:
    def __init__(self, nama, hp, attack):
        self.nama = nama
        self.attack = attack
        # 2. ENKAPSULASI (Brankas Rahasia): 
        # Kita pakai tanda __ di depan hp agar nyawa tidak bisa di-cheat dari luar program!
        self.__hp = hp 

    # Karena __hp digembok, kita harus buat "jalur resmi" untuk melihat nyawa
    def lihat_hp(self):
        return self.__hp

    # "Jalur resmi" untuk mengurangi/menambah nyawa kalau kena pukul atau minum potion
    def atur_hp(self, jumlah):
        self.__hp += jumlah
        if self.__hp <= 0:
            self.__hp = 0 # Biar nyawa tidak minus
            print(f"💀 {self.nama} telah K.O!")

    # Aksi standar karakter
    def status(self):
        print(f"[{self.nama}] HP: {self.__hp}")


# --- 3. INHERITANCE (Pewarisan) ---
# Hero adalah 'Anak' dari Karakter. Dia mewarisi nama, HP, dan attack.
class Hero(Karakter):
    
    # Skill khusus Hero (Monster tidak punya ini)
    def minum_potion(self):
        print(f"✨ {self.nama} meminum Potion! HP bertambah 20.")
        self.atur_hp(20) # Lewat jalur resmi nambah nyawa

    # 4. POLIMORFISME (Beda Gaya)
    # Hero punya gaya menyerang yang beda, dia pakai pedang dan ada peluang Critical Hit!
    def serang(self, target):
        damage = self.attack
        is_critical = random.choice([True, False]) # Peluang 50:50
        
        if is_critical:
            damage *= 2
            print(f"💥 CRITICAL HIT! {self.nama} menebas {target.nama} dengan damage {damage}!")
        else:
            print(f"⚔️ {self.nama} menebas {target.nama} dengan damage {damage}!")
        
        target.atur_hp(-damage) # Mengurangi nyawa target


# --- 3. INHERITANCE (Pewarisan) ---
# Monster juga 'Anak' dari Karakter.
class Monster(Karakter):
    
    # 4. POLIMORFISME (Beda Gaya)
    # Gaya serang monster beda dengan hero. Dia menyemburkan api!
    def serang(self, target):
        print(f"👾 {self.nama} menyemburkan api ke {target.nama}! Damage: {self.attack}")
        target.atur_hp(-self.attack)


# ==========================================
# --- MARI KITA MAIN DI LAPANGAN (OBJEK) ---
# ==========================================

print("=== PERTARUNGAN DIMULAI ===")
# Kita cetak OBJEK aslinya!
arthur = Hero("Arthur si Kesatria", 100, 15)
naga = Monster("Naga Hitam", 80, 10)

# Cek status awal
arthur.status()
naga.status()
print("-" * 30)

# Ronde 1: Arthur menyerang Naga
arthur.serang(naga)
# Ronde 2: Naga membalas
naga.serang(arthur)
print("-" * 30)

# Arthur panik, nyawanya berkurang. Dia minum potion!
arthur.minum_potion()
print("-" * 30)

# Cek status terakhir
arthur.status()
naga.status()

# -- TES ENKAPSULASI (ANTI CHEAT) --
# Coba hapus tanda pagar (#) pada baris di bawah ini dan jalankan ulang untuk melihat apa yang terjadi:
# arthur.__hp = 9999
```
